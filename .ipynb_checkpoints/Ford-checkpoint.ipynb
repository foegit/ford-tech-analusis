На сайті Alpha Vantage потрібно отримати спеціальний *api-ключ* за допомогою якого буде здійснюватись доступ до даних. Також потрібно знайти ярлик (symbol) компанії Ford.

Опишемо функцію для отримання даних з сайту Alpha Vantage.

```python
def get_alpha_vantage(
        apikey, symbol,
        function='TIME_SERIES_DAILY',
        outputsize='compact',
        datatype='csv',
        **kwargs):

    '''Makes a request with Alpha Vantage API and returns a pandas Dataframe. 

    Usage:
    apple = get_alpha_vantage(apikey, 'AAPL')
    intel = get_alpha_vantage(apikey, 'INTC', outputsize='full')
    bitcoin = get_alpha_vantage(
        apikey, 'BTC', function='DIGITAL_CURRENCY_DAILY',
        outputsize='compact', market='USD')
    
    '''

    base_url = 'https://www.alphavantage.co/query?'
    request_url = base_url + '&'.join\
        (['function=' + function,
          'outputsize=' + outputsize,
          'datatype=' + datatype,
          'symbol=' + symbol,
          'apikey=' + apikey])
    
    for key, val in kwargs.items():
        request_url += '&' + str(key) + '=' + val

    data = pd.read_csv(request_url, index_col='timestamp', parse_dates=True)    
    return data.iloc[::-1]
```

Щоб отримати дані достатньо викликати данну функції з відповіднми параметрами.

```python
ford_data = get_alpha_vantage(api_key, ford_symbol, outputsize='full')
print(ford_data)
```

Ми успішно отримали дані про ціни компанії Ford в період з 1998 до сьогоднішнього дня. Для візуального аналізу побудуємо графік.

```python
ford_data['close'].plot(title = '{} price'.format('Ford'))
```

На графіку видно, що ціни акцій Ford стабільно падають з 2014 року. Тому для аналізу візьмемо саме цей період.

```python
ford_data = ford_data['2014':]
ford_data['close'].plot(title = '{} price'.format('Ford'))
```

З 2014 року ціни на акції форм стабільно падають, але графік доволі отнотипний, а ріст і падіння коливаються відносно сильно. Опишемо функцію яка рисує японські свічки для наших данних.

```python
def plot_candlestick(data):    
    ax = plt.axes()
    candlestick2_ohlc(
            ax,
            data['open'], data['high'], data['low'], data['close'],
            width=1)
```

```python
plot_candlestick(ford_data['2019':])
```
